```python
import labtools as lt 
import numpy as np 
import pandas as pd 
from matplotlib import pyplot as plt 
from matplotlib import cm
import os
from labtools import scanner
import seaborn as sns
from scipy.optimize import curve_fit
from scipy.signal import find_peaks
from labtools import plot_style
```

```python
from ipywidgets import RadioButtons, Layout
from IPython.display import display
from ipywidgets import HBox, VBox
from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widgets
```

```python
folder = "/home/bbuusshh/my_git/data/fiber/pressure2"
```

```python
scans = pd.DataFrame(os.listdir(os.path.join(folder)), columns=['Name'])

scans = scans.iloc[np.argsort(scans["Name"].apply(lambda x: int(x[-2:])))]

imgs = []
headers = []
freqs = []
for scan in scans["Name"]:
    header, raw, freq = scanner.xtract(f"{folder}/{scan}")
    imgs.append(raw)
    headers.append(header)
    freqs.append(freq)
```

```python
def plot_imgs(scan):
    plt.imshow(imgs[scan], aspect='auto')
    
interact(
    plot_imgs, 
    scan = widgets.RadioButtons(
    options=range(len(scans)),
    description='Scans:',
    disabled=False,
))
```

```python
im1 = pd.DataFrame(imgs[11]/1000, columns=(freqs[11] - freqs[11][0]).astype(int)).T

time = (im1.columns * 157*13e-3/60).astype(int)

im1.columns = time
```

```python

plt.figure(figsize=(10, 8))
sns.set(font_scale=2.3, style='ticks', font="Arial")
ax = sns.heatmap(im1,yticklabels=25,xticklabels=150, cmap=cm.viridis)
ax.set(xlabel='time, min', ylabel=r'$\Delta\nu$, MHz')

plt.yticks(rotation=0) 
plt.tight_layout()
plt.savefig("pressure_discharge.png", dpi=300)
```
