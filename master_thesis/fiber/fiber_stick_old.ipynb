```python
def correlation3(data, corr_window=600_000, ch1=[1], ch2=[2],ch3=[3], trig_ch=5):
    autoc_data = data[data['channel'] != 5]
    autoc_diffs_1 = []
    autoc_diffs_2 = []
    autoc_diffs_3 = []
    
    autoc_diffs_t1 = []
    autoc_diffs_t2 = []
    autoc_diffs_t3 = []
    
    
    shift_1=0
    shift_2=0
    while True:#shift != 20:
        shift_1 += 1
        shift_2 += 2
        diffs_1 = autoc_data['time'] - np.roll(autoc_data['time'], shift_1)
        diffs_2 = autoc_data['time'] - np.roll(autoc_data['time'], shift_2)
        dif_ch_1 = np.roll(autoc_data['channel'], shift_1)
        dif_ch_2 = np.roll(autoc_data['channel'], shift_2)
        
        diffs_1a = diffs_1[(diffs_1 < corr_window) & np.isin(autoc_data['channel'], ch1) & np.isin(dif_ch_1, [ch2, ch3])]
        diffs_1b = diffs_1[(diffs_1 < corr_window) & np.isin(autoc_data['channel'], [ch2, ch3]) & np.isin(dif_ch_1, ch1)]
        autoc_diffs_1.append(diffs_1a.astype('int32'))
        autoc_diffs_1.append(-diffs_1b.astype('int32'))
        
        diffs_2a = diffs_1[(diffs_1 < corr_window) & np.isin(autoc_data['channel'], ch2) & np.isin(dif_ch_1, [ch1, ch3])]
        diffs_2b = diffs_1[(diffs_1 < corr_window) & np.isin(autoc_data['channel'], [ch1, ch3]) & np.isin(dif_ch_1, ch2)]
        autoc_diffs_2.append(diffs_2a.astype('int32'))
        autoc_diffs_2.append(-diffs_2b.astype('int32'))
        
        diffs_3a = diffs_1[(diffs_1 < corr_window) & np.isin(autoc_data['channel'], ch3) & np.isin(dif_ch_1, [ch1, ch2])]
        diffs_3b = diffs_1[(diffs_1 < corr_window) & np.isin(autoc_data['channel'], [ch1, ch2]) & np.isin(dif_ch_1, ch3)]
        autoc_diffs_3.append(diffs_3a.astype('int32'))
        autoc_diffs_3.append(-diffs_3b.astype('int32'))
        
        diffs_t1a = diffs_2[(diffs_2 < corr_window) & np.isin(autoc_data['channel'], ch1) & np.isin(dif_ch_2, [ch2, ch3])]
        diffs_t1b = diffs_2[(diffs_2 < corr_window) & np.isin(autoc_data['channel'], [ch2, ch3]) & np.isin(dif_ch_2, ch1)]
        autoc_diffs_t1.append(diffs_t1a.astype('int32'))
        autoc_diffs_t1.append(-diffs_t1b.astype('int32'))
        
        diffs_t2a = diffs_2[(diffs_2 < corr_window) & np.isin(autoc_data['channel'], ch2) & np.isin(dif_ch_2, [ch1, ch3])]
        diffs_t2b = diffs_2[(diffs_2 < corr_window) & np.isin(autoc_data['channel'], [ch1, ch3]) & np.isin(dif_ch_2, ch2)]
        autoc_diffs_t2.append(diffs_t2a.astype('int32'))
        autoc_diffs_t2.append(-diffs_t2b.astype('int32'))
        
        diffs_t3a = diffs_2[(diffs_2 < corr_window) & np.isin(autoc_data['channel'], ch3) & np.isin(dif_ch_2, [ch1, ch2])]
        diffs_t3b = diffs_2[(diffs_2 < corr_window) & np.isin(autoc_data['channel'], [ch1, ch2]) & np.isin(dif_ch_2, ch3)]
        autoc_diffs_t3.append(diffs_t3a.astype('int32'))
        autoc_diffs_t3.append(-diffs_t3b.astype('int32'))
        
        lensT = len(diffs_t3a)*len(diffs_t3b)*len(diffs_t2a)*len(diffs_t2b)*len(diffs_t1a)*len(diffs_t1b)
        lens =  len(diffs_1a)*len(diffs_1b)*len(diffs_2a)*len(diffs_2b)*len(diffs_3a)*len(diffs_3b)
        if lensT*lensT == 0:
            break
#     autoc_diffs_12 = np.array(autoc_diffs_12)
#     autoc_diffs_23 = np.array(autoc_diffs_23)
#     autoc_diffs_31 = np.array(autoc_diffs_31)
#     autoc_diffs_tot_12 = np.concatenate(autoc_diffs_12)
#     autoc_diffs_tot_23 = np.concatenate(autoc_diffs_23)
#     autoc_diffs_tot_31 = np.concatenate(autoc_diffs_31)
    return autoc_diffs_1,autoc_diffs_2,autoc_diffs_3,autoc_diffs_t1,autoc_diffs_t2,autoc_diffs_t3
#autoc_diffs_tot_12,autoc_diffs_tot_23 ,autoc_diffs_tot_31, autoc_diffs_12, autoc_diffs_23, autoc_diffs_31
```

```python
def correlation3(data, corr_window=30000, ch1=1, ch2=2, ch3=3, trig_ch=5, binNum=100):
    binWidth = corr_window // binNum
    data = data[data['channel'] != 5]
    autoc_diffs = []
    shift_1,shift_2=0,0
    g3 = []
    hist2d = np.zeros((binNum, binNum))
    
    shift_1 = 1
    data_r1 = np.roll(data, shift_1)
    roll_1 = data_r1.copy()
    roll_1['time'] = data_r1['time'] - roll_1['time']
    chan_cond_1 = (np.isin(data['channel'], ch1) & np.isin(roll_1['channel'], [ch2, ch3]))\
    | (np.isin(data['channel'], ch2) & np.isin(roll_1['channel'], [ch1, ch3]))\
    | (np.isin(data['channel'], ch3) & np.isin(roll_1['channel'], [ch2, ch1]))
    diffs_1 = roll_1['time'][(roll_1['time'] < corr_window) & chan_cond_1]
    roll_1, data = None, None
    while True:#shift != 20:
        shift_2 += 1
        data_r2 = np.roll(data_r1, shift_2)
        roll_2 = data_r2.copy()
        roll_2['time'] = data_r1['time'] - roll_2['time']
        chan_cond_2 =(np.isin(data_r1['channel'], ch1) & np.isin(roll_2['channel'], [ch2, ch3]))\
    | (np.isin(data_r1['channel'], ch2) & np.isin(roll_2['channel'], [ch1, ch3]))\
    | (np.isin(data_r1['channel'], ch3) & np.isin(roll_2['channel'], [ch2, ch1]))

        diffs_2 = roll_2['time'][(roll_2['time'] < corr_window) & chan_cond_2]

        y_p,x = np.histogram(diffs_2, bins = binNum, range=(0, corr_window))
        binInds = diffs_1 // binWidth
        hist2d[binInds] += y_p



        if (len(diffs_2) == 0):
            break
    return hist2d

```
