```python
import labtools
```

```python
PARAMS = {
    "title_font_size": 25,
    "grid_linewidth": 1,
    "grid_color":"orange",
    "grid_linestyle": "--",
    "ticks_lenght": 4,
    "ticks_lenght_minor": 2.5,
    "ticks_width": 2,
    "grid_alpha": 0.4,
    "labels_font_size": 35,
    "saveDir": ".",
    "fontsize":25,
    "fontStyle":'Helvetica',
    "plot_name":'default_name'
}
```

```python
from labtools import plot_style
```

```python
p = plot_style.plotting(PARAMS)
p.saveDir = "/Users/vladislavbushmakin/Desktop/presa_pics"
```

```python
from matplotlib import pyplot as plt
```

```python
nodes = range(21)
effs = [0.85, 0.9, 0.95, 0.99]
rate = [effs[0]**i for i in nodes]
p.plot(nodes, rate, xLabel="nodes", yLabel='rate fraction', tick_step_x=5)
for eff in effs[1:]:
    rate = [eff**i for i in nodes]
    plt.plot(nodes, rate)
p.save(f'all')
```
